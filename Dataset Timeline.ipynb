```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('data/0f_turkey.csv')
```

```python
print(df1.columns)
```

```python
df1['date'] = pd.to_datetime(df1['date'], format='%Y-%m-%d %H:%M:%S%z')
df1.set_index('date', inplace=True)

tweets_per_minute1 = df1.resample('T').size()
tweets_per_hour1 = df1.resample('H').size()
tweets_per_day1 = df1.resample('D').size()

fig, axs = plt.subplots(3, 1, figsize=(12, 10))

axs[0].plot(tweets_per_minute)
axs[0].set_title('Turkey : Tweets per Minute')
axs[0].set_xlabel('Time')
axs[0].set_ylabel('Number of Tweets')
axs[0].grid(True)

axs[1].plot(tweets_per_hour)
axs[1].set_title('Turkey : Tweets per Hour')
axs[1].set_xlabel('Time')
axs[1].set_ylabel('Number of Tweets')
axs[1].grid(True)

axs[2].plot(tweets_per_day)
axs[2].set_title('Turkey : Tweets per Day')
axs[2].set_xlabel('Time')
axs[2].set_ylabel('Number of Tweets')
axs[2].grid(True)

plt.tight_layout()
plt.show()
```
